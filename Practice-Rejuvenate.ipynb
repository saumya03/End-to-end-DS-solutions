```python
dff = pd.read_csv("SpaData.csv")
```

```python
dff.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Let's say we want to split the data in 80:10:10 for train:valid:test dataset
train_size=0.8

X = dff.drop(columns = ['GTOccupancy', 'TimeStamp'], axis=1).copy()
y = dff['GTOccupancy']
```

```python
# In the first step we will split the data in training and remaining dataset
X_train, X_rem, y_train, y_rem = train_test_split(X,y, train_size=0.8) #training data
```

```python
# Now since we want the valid and test size to be equal (10% each of overall data). 
# we have to define valid_size=0.5 (that is 50% of remaining data)
test_size = 0.5
X_valid, X_test, y_valid, y_test = train_test_split(X_rem,y_rem, test_size=0.5)
```

```python
print(X_train.shape), print(y_train.shape)
print(X_valid.shape), print(y_valid.shape)
print(X_test.shape), print(y_test.shape)
```

```python
#KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
#find optimum no. of neighbors, in b/w 1 to 50 using grid search
hyper_params_space = [{'n_neighbors': np.arange(1, 50),},]
knn = GridSearchCV(KNeighborsClassifier(), hyper_params_space,
                       scoring='accuracy')
knn.fit(X_train, y_train)
    
print('Best parameters set:')
print(knn.best_params_)
print()
```

```python
#elbow method
error_rate = []
# Will take some time
for i in range(1,40):
 
 knn = KNeighborsClassifier(n_neighbors=i)
 knn.fit(X_train,y_train)
 pred_i = knn.predict(X_test)
 error_rate.append(np.mean(pred_i != y_test))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color= "blue", linestyle="dashed", marker="o",
 markerfacecolor="red", markersize=10)
plt.title("Error Rate vs. K Value")
plt.xlabel("K")
plt.ylabel("Error Rate")
```

```python
classifier = KNeighborsClassifier(n_neighbors = 13)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(result)
result1 = classification_report(y_test, y_pred)
print("Classification Report:",)
print (result1)
result2 = accuracy_score(y_test,y_pred)
print("Accuracy:",result2)
```
